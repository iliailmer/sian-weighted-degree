```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("../systems/cholera_degrees_per_var.json", "r") as f:
    cholera = json.load(f)

with open("../systems/cholera_degrees_per_var_subs.json", "r") as f:
    cholera_subs = json.load(f)
    
with open("../systems/cholera_degrees_per_var_old_subs.json", "r") as f:
    cholera_old_subs = json.load(f)
    
with open("../systems/cholera_degrees_per_var_old_subs_2.json", "r") as f:
    cholera_old_subs_2 = json.load(f)

cholera_degree_matrix = np.array(list(map(json.loads, cholera.values())))
cholera_subs_degree_matrix = np.array(list(map(json.loads, cholera_subs.values())))
cholera_old_subs_degree_matrix = np.array(list(map(json.loads, cholera_old_subs.values())))
cholera_old_subs_2_degree_matrix = np.array(list(map(json.loads, cholera_old_subs_2.values())))


```

```python
fig = plt.figure(figsize=(12, 8))
plt.imshow(cholera_degree_matrix)
```

```python
fig = plt.figure(figsize=(12, 8))
plt.imshow(cholera_subs_degree_matrix)
```

```python
fig = plt.figure(figsize=(12, 8))
plt.imshow(cholera_old_subs_degree_matrix)
```

```python
fig = plt.figure(figsize=(12, 8))
plt.imshow(cholera_old_subs_2_degree_matrix)
```
